```python
import os
import random
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt

# Load the YOLO model
model = YOLO('/content/drive/MyDrive/Colab Notebooks/YOLOv11 ECG Dataset/saved_model.pt', task='detect')

# Path to the directory containing test images
test_images_path = '/content/drive/MyDrive/Colab Notebooks/YOLOv11 ECG Dataset/Dataset/dataset/test/images'

# Get a list of all image files in the directory
image_files = [f for f in os.listdir(test_images_path) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Select a number of random images (e.g., 5 images)
num_random_images = 5
random_images = random.sample(image_files, num_random_images)

# Get the full paths of the selected random images
random_image_paths = [os.path.join(test_images_path, img) for img in random_images]

# Print the selected images
print(f"Randomly selected images: {random_images}")

# Perform prediction and display the results for each selected image
for image_path in random_image_paths:
    # Perform prediction
    results = model.predict(image_path, save=True)  # Saves the image with bounding boxes

    # Load the saved image with predictions
    result_image_path = results[0].path  # This gets the path of the saved image with predictions
    predicted_image = cv2.imread(result_image_path)
    predicted_image = cv2.cvtColor(predicted_image, cv2.COLOR_BGR2RGB)  # Convert to RGB for display

    # Get prediction details (boxes, class IDs, etc.)
    boxes = results[0].boxes.xyxy  # Bounding boxes (xyxy format)
    class_ids = results[0].boxes.cls  # Class IDs for the predictions
    confidences = results[0].boxes.conf  # Confidence scores for each prediction
    class_names = model.names  # Get the class names

    # Loop through each prediction and draw the bounding box with class name
    for box, class_id, confidence in zip(boxes, class_ids, confidences):
        x1, y1, x2, y2 = map(int, box)  # Get the coordinates of the bounding box
        class_name = class_names[int(class_id)]  # Get the class name
        label = f"{class_name} ({confidence:.2f})"  # Prepare label with class and confidence score

        # Draw the bounding box and label on the image
        cv2.rectangle(predicted_image, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(predicted_image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Display the image with predictions
    plt.figure(figsize=(6, 6))
    plt.imshow(predicted_image)
    plt.axis('off')
    plt.show()

    # Optionally, print the results (predictions)
    print(f"Prediction results for {image_path}:")
    print(results)
```
